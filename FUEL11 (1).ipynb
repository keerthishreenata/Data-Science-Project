```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Check the distribution of the target variable (Y)
value_counts = pd.Series(Y).value_counts()

# Identify classes with fewer than 2 samples
classes_to_remove = value_counts[value_counts < 2].index

# Remove these classes from the dataset
X_filtered = X[~Y.isin(classes_to_remove)]
Y_filtered = Y[~Y.isin(classes_to_remove)]

# Perform the train-test split on the filtered data
X_train, X_test, Y_train, Y_test = train_test_split(X_filtered, Y_filtered, test_size=0.2, stratify=Y_filtered, random_state=2)

# Check the distribution of the target variable after filtering
print(pd.Series(Y_filtered).value_counts())
```

```python
# splitting
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming Y is your target variable (continuous)
# Convert continuous target variable into discrete bins (e.g., low, medium, high)
Y_binned = pd.cut(Y, bins=3, labels=['low', 'medium', 'high'])

# Perform stratified train-test split based on binned labels
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_binned, test_size=0.2, stratify=Y_binned, random_state=2)

# Check the distribution of binned target in training and testing sets
print("Training set target distribution:")
print(Y_train.value_counts())

print("Testing set target distribution:")
print(Y_test.value_counts())
```
